## Colored Images

**1 pixel like 1 row and 3 col's**

```python
# 3 dimensional since [[[]]], 3 brackets i.e. (x,y,z)
# 1 layer since only 1 double bracket [[]]
# 1 row since only 1 single bracket inside [[]]
# 3 columns since 3 elements seperated by comma within single bracket []

arr3 = np.array([[[255,0,0]]])   
print(arr3.shape)       # 1 layer, 1 row, 3 columns
plt.imshow(arr3);
```

```python
# image properties
print("Type:",type(arr3))   
print("Shape of Image:", arr3.shape)   
print('Total Number of pixels:', arr3.size)  
print("Image data type:", arr3.dtype)

print("Pixel Values:\n", arr3)
print("Dimension:", arr3.ndim) 
```

```python
arr4 = np.array([[[255,0,0],[0,0,255]]]) #  1 layer since one [[]]
                                #  2 rows as two [] seperated by comma
print(arr4.shape)# 3 cols as within each [], 3 elements seperated by comma
plt.imshow(arr4);       
```

```python
# image properties
print("Type:",type(arr4))   
print("Shape of Image:", arr4.shape)   
print('Total Number of pixels:', arr4.size)  
print("Image data type:", arr4.dtype)

print("Pixel Values:\n", arr4)
print("Dimension:", arr4.ndim)  
```

```python
arr5 = np.array([[[255,0,0]],    # 2 layers as two [[]] seperated by comma
                 [[0,0,255]]])   
print(arr5.shape)     # 1 row as no single bracket [] seperated by comma 
plt.imshow(arr5);     # 3 cols as within [], 3 elements seperated by comma
```

```python
# image properties
print("Type:",type(arr5))   
print("Shape of Image:", arr5.shape)   
print('Total Number of pixels:', arr5.size)  
print("Image data type:", arr5.dtype)

print("Pixel Values:\n", arr5)
print("Dimension:", arr5.ndim) 
```

## Extraction of rows & columns/each 2D surface in 3D

## <font color=blue>**Note on Interpretation of 3D matrix**

**- If biggest bracket is [[[]]], it is 3 dimensional i.e shape ( * , * ,*  ) which is (layers,rows,cols)**


**- Number of components/layers is denoted by number of double brackets[[]],if 3 double brackets, so shape (3,rows ,columns)**


**- Within each layer where each layer is represented by [[]], the number of rows will be equal to number of single brackets [] which are seperated by comma**


**- Number of columns will be equal to number of elements within each single bracket [] which are seperated by comma.**

```python
[[[]],           # 3 dimensional represented by [[[]]]
 [[]],           # layers/components represent by [[]], here 3
 [[]]]           # [] seperated by comma will represent rows, here only 1
                 
```

```python
a = np.array([[[2]],        # 3 dimensional [[[]]]
              [[3]],        # layers represent by [[]], here 3
              [[4]]])       #  Within each 2D plane, rows defined by       
print(a.shape)              # [] seperated by comma, here only 1

print(a)
```

```python
[[[],[]],      # 3 dimensional
 [[],[]],      # planes/layers/components represent by [[]] in 3-D, here 3
 [[],[]]]      # Within each [[]], rows represent by [] seperated by ,
               # shape = (3,2,1)  there are three 2 * 1 matrix
```

```python
b = np.array([[[1],[2]], # 3 dimensional with 3 layers as 3 [[]]
              [[6],[3]], # rows represent by [] seperated by , within [[]]
              [[4],[4]]])   #  cols is 1 as only 1 element inside each [] 

print(b.shape)      # there are three, 2 * 1 matrix
print(b)
                    
```

```python
c = np.array([[[3,4,5]],    # 3 layers as 3 double brackets [[]]   
              [[7,8,9]],    # Within each layer[[]], [] seperated by ,
              [[4,9,0]]])   # show rows here no [] seperated by , so 1

                            # Within each [], elements seperated by ,
                            # show columns, here 3 elements seperated by
                            # by comma within single bracket
print(c.shape)
c
```

```python
d = np.array([[[4,5],[8,9]],
              [[7,8],[0,8]]])  # 2 layers, each layer has 2rows* 2cols
print(d.shape)
d
```

## Read & Display image (by default BGR)

```python
os.chdir(r"/Users/sylvia/Desktop/datasets/CV")
img = cv2.imread('multi.jpg', 1)      # By default BGR format
plt.imshow(img);
```

```python
def resize(img, width, height,ip):
    dim = (width, height)
    resize_img = cv2.resize(img, dim, interpolation=ip)
    return resize_img
```

```python
resize_img = resize(img, 3,3,cv2.INTER_AREA)

print('original image:', img.shape)
print('resized image:', resize_img.shape)

plt.imshow(resize_img);
```

```python
resize_img      ## 0th column is Blue, 1st col is Green, 2nd col is red
```

```python
# Access 1st plane/layer of 3D matrix which is first double bracket [[]]
p1 = resize_img[0,:,:]
print(p1.shape)
p1
```

```python
# Access 2nd plane/layer of 3D matrix which is second double bracket [[]]
p2 = resize_img[1,:,:]
print(p2.shape)
p2
```

```python
# Access 3rd plane/layer of 3D matrix which is third double bracket [[]]
p3 = resize_img[2,:,:]
print(p3.shape)
p3
```

```python
# Access individual pixels of 3D array
```

```python
# 1st layer i.e 1st 2D plane, index 0
print(resize_img[0,0,:])     # 1st pixel (1row,3cols make 1 pixel)
print(resize_img[0,1,:])     # 2nd pixel
print(resize_img[0,2,:])     # 3rd pixel
```

```python
# 2nd layer i.e 2nd 2D plane
# [[223,157,81], [179,136,96], [132,117,100]]
```

```python
# 3rd layer i.e 3rd 2D plane, index 2

print(resize_img[2,0,:])    # 7th pixel
print(resize_img[2,1,:])    # 8th pixel
print(resize_img[2,2,:])    # 9th pixel
```

```python
# 3rd layer i.e 3rd 2D plane
# [[62,61,213],  [47,53,220],  [70,73,169]]
```
