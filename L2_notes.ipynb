### Relu Function

<font color=blue>The Rectified Linear Unit (ReLU) function, defined as \( f(x) = \max(0, x) \), is widely used as an activation function in deep neural networks (mostly used in intermediate layers) for several reasons:

### 1. Computational Efficiency:
ReLU is computationally inexpensive to evaluate. This can significantly speed up the training process compared to activation functions like the sigmoid or tanh that involve more complex mathematical operations (e.g., exponentials).

### 2. Sparsity:
ReLU activation leads to sparsity. When the output is zero, it's essentially ignoring that particular neuron, leading to a sparse representation. Sparsity is beneficial because it makes the network easier to optimize and can lead to a more efficient model.

### 3. Gradient Propagation:
For positive inputs, the gradient of the ReLU function is 1, so during backpropagation, it effectively allows the gradient to pass through unchanged for those neurons that are activated. This mitigates the vanishing gradient problem, which is when gradients become too small for the network to learn effectively. The vanishing gradient problem is particularly troublesome for deep networks.

### 4. Non-linearity:
Even though it looks like a linear function, ReLU introduces non-linearity into the network, allowing it to learn from the error and make adjustments. This is essential for learning complex patterns.

### 5. Simplicity:
ReLU is simple to implement and is often the default choice for many types of neural networks.

### 6. Biological Interpretations:
Some evidence suggests that ReLU units are more biologically plausible than sigmoidal units as models of biological neurons. This is, however, still a topic of ongoing research.

### Drawbacks and Variants:
It's worth mentioning that ReLU is not without its problems. The most notable one is the "dying ReLU" problem, where neurons can sometimes get stuck during training and always output zero. This can happen if a large gradient flows through a ReLU neuron, updating the weights in such a way that the neuron will always output zero. Variants like the Leaky ReLU or Parametric ReLU have been proposed to address this issue. These functions allow a small, non-zero gradient when the unit is not active and are defined as \( f(x) = \alpha x \) for \( x < 0 \), \( f(x) = x \) for \( x \geq 0 \), where \( \alpha \) is a small constant.

Despite its drawbacks, the advantages of ReLU make it a popular choice for many types of neural network architectures.

### Cost vs Objective function

So while both are optimization targets, the term "cost function" usually implies something you want to minimize, and "objective function" is more general, describing what you want to optimize without specifying direction (minimize or maximize).






Cost Function: Primarily used in the context of supervised learning, focuses on minimizing the discrepancy between predicted and actual labels.

Objective Function: A more general term, used in various types of optimization problems, that could involve either minimization or maximization.

### What makes 1 algorithm different from another inspite of same loss function which we r trying to minimize?

So, while the cost function serves the purpose of guiding the model during the training phase, what makes one model different from another extends beyond the cost function to include architectural complexity, parameter count, flexibility, and the types of relationships they can capture.

### Parameters

1. **Linear Regression**: The number of parameters corresponds to the number of features plus an optional bias term. All these parameters are directly involved in predicting the output.
  
2. **Neural Networks**: The number of parameters can be vastly larger and are spread across multiple layers. These include weights and biases at each layer and for each neuron.

### Architecture

- **Linear Regression**: This model has a straightforward architecture, assuming a linear relationship between inputs and outputs.

- **Neural Networks**: The architecture can be quite complex, involving multiple layers, various types of layers (convolutional, recurrent, attention, etc.), and different connections between layers.

### Non-Linearity

- **Linear Regression**: No inherent ability to model non-linear relationships unless you engineer non-linear features manually.
  
- **Neural Networks**: They incorporate non-linearity through activation functions, allowing the model to learn from the error and adjust its weights to model non-linear relationships effectively.

### Flexibility

- **Linear Regression**: Less flexible due to its simplicity.

- **Neural Networks**: More flexible and can adapt to a variety of data types and complexities.

### Regularization and Optimization

- Both models can employ different regularization techniques. Linear models often use L1 (Lasso) and L2 (Ridge) regularization, while neural networks might use dropout, batch normalization, and early stopping.

- The optimization landscape for neural networks is generally much more complicated, leading to the use of more advanced optimization techniques beyond basic gradient descent.

### Interpretability

- **Linear Regression**: Easy to interpret the model parameters in terms of their impact on the output.

- **Neural Networks**: Generally considered "black boxes" due to their complexity and the large number of parameters.







### Backpropagation and Gradient Descent

Backpropagation and gradient descent are related but not the same. In a nutshell:

- **Backpropagation** is an algorithm for efficiently computing gradients of the loss function with respect to the weights of the network, effectively applying the chain rule of calculus.
  
- **Gradient Descent** is an optimization algorithm used to minimize some function by iteratively moving in the direction of steepest decrease, defined by the negative of the gradient.

In the context of neural networks, backpropagation provides the gradients of the loss function with respect to each parameter (weight and bias), and gradient descent uses these gradients to update the parameters and minimize the loss. Backpropagation is the "how to compute the gradients" part, while gradient descent is the "what to do with the gradients once you have them" part.

### Activation function in intermediate layers - Relu

We use Relu in intermediate layers. For any non-linearity to be brought about we use Relu we do not want to leave our function linear. Also relu will bring about faster convergence i.e. to reach the minimum of loss function. Sigmoid and Softmax are for a reason used in output layer.

Relu to be used in intermediate layers is a proven fact by researchers and we follow it here.

The choice of activation functions in neural networks depends on various factors such as the problem being solved, the nature of the input and output data, and the network architecture. Here's a general idea of why ReLU (Rectified Linear Unit) is commonly used in intermediate layers, while sigmoid and softmax are frequently employed in the output layer:

### ReLU in Intermediate Layers:

1. **Computational Efficiency**: ReLU is computationally less expensive compared to sigmoid and tanh because it only involves a simple thresholding at zero. This allows models to train faster and requires less computational resources compared to other activation functions.

2. **Sparsity**: When the output is zero, it's effectively ignoring that neuron, leading to a sparse representation. Sparsity is beneficial because it makes the network easier to optimize and can lead to a more efficient model.

3. **Mitigating the Vanishing Gradient Problem**: ReLU activation leads to a gradient of either zero (for negative inputs) or one (for positive inputs), which helps mitigate the vanishing gradient problem, a situation where gradients become too small for effective training in deep networks.

4. **Non-linearity**: Despite its simplicity, ReLU introduces non-linearity into the network, allowing it to learn from the error and make adjustments, which is essential for learning complex patterns.

### Sigmoid and Softmax in Output Layers:

1. **Probability Estimation**: The sigmoid function maps any input into a value between 0 and 1, making it useful for binary classification problems where you need a probability estimate as output.

2. **Multi-class Classification**: Softmax is used in the output layer for multi-class classification problems. It converts the raw output into probability distribution over multiple classes.

3. **Differentiable**: Both sigmoid and softmax functions are smooth and differentiable. This is useful for gradient-based optimization methods.

4. **Bound Output**: Sigmoid and softmax both provide output in a specific range (0-1), which can be useful in scenarios where the output needs to be normalized.


- **ReLU** is commonly used in intermediate layers due to its computational efficiency, sparsity, and ability to mitigate the vanishing gradient problem.
  
- **Sigmoid and Softmax** are commonly used in the output layer to produce probabilities and are particularly useful for classification problems.
