## <font color=blue>Why min-max scaling is used in case of images instead of standardization?

Both Min-Max normalization and Standardization (Z-score normalization) are common techniques for feature scaling, but they serve different purposes and are useful in different contexts. In image processing, Min-Max normalization is often preferred for a few reasons:

### Pixel Intensity Range:
1. **Bounded Range**: Pixel intensities in an image are often bounded within a specific range. For 8-bit grayscale images, the pixel values range from 0 to 255. Min-Max normalization scales these values to a different bounded range, such as [0, 1], which is easier to work with in neural networks.

2. **Non-Negative Values**: Min-Max normalization ensures that all pixel values are non-negative, which is important for many image processing algorithms and neural network architectures.

3. **Preservation of Relationships**: Min-Max normalization preserves the relationship between the original pixel values, which is important for capturing the structural and textural information in the image.

### Computational Reasons:
4. **Numerical Stability**: When working with neural networks, especially deep learning models like CNNs (Convolutional Neural Networks), having input features like image pixels in a bounded, consistent range can help with the numerical stability of the training process.

5. **Activation Functions**: Some activation functions like the sigmoid or tanh are more sensitive to the scale of the input. Normalizing image pixel values can help these functions operate in their "active" range.

### Interpretability:
6. **Human Interpretability**: Min-Max normalization makes the data more understandable and easier to visualize, especially when transformed into images. This can be useful in debugging or in the interpretation of the model outputs.

7. **Simplicity**: Min-Max is straightforward and easy to implement. For images, where pixel intensities do not usually follow a Gaussian distribution, this method can be more appropriate.

Standardization, which transforms data to have zero mean and unit variance, is generally more useful when the features have a Gaussian (normal) distribution and they differ in ranges. However, this isn't often the case with images, making Min-Max a more natural choice for image data.

That said, the choice between Min-Max normalization and Standardization isn't strict, and depending on the specific application, dataset, and model architecture, Standardization could also be used.

## <font color=blue>Updation of weights in neural network

In machine learning, particularly in neural networks, the way weights are updated can vary depending on the training configuration. Here are the main approaches:

1. **Batch Gradient Descent**: In this approach, weights are updated after processing the entire dataset, i.e., after every epoch. You calculate the gradient using the entire dataset and update the weights.

2. **Stochastic Gradient Descent (SGD)**: In this case, weights are updated after every single data point, i.e., after every batch where the batch size is 1. This can lead to more noisy weight updates, but sometimes that noise can help escape local minima.

3. **Mini-batch Gradient Descent**: This is a compromise between Batch Gradient Descent and Stochastic Gradient Descent. In Mini-batch Gradient Descent, weights are updated after processing a mini-batch of `n` samples, where `1 < n < size of the dataset`. This is the most commonly used approach in practice.

The terms "epoch" and "batch" can be defined as:

- **Epoch**: One complete forward and backward pass of all the training examples.
- **Batch**: A set of `N` training examples, where `N` can vary from 1 to the size of the training set. The number of batches in one epoch is `ceil(total number of samples / batch size)`.

So to directly answer your question: weights can be updated either after every epoch or after every batch within an epoch, depending on the specific variant of gradient descent being used. Most commonly in practice, weights are updated after every mini-batch.
