```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense,Conv2D,MaxPooling2D
```

```python
model=Sequential()
model.add(Conv2D(filters=6,kernel_size=(5,5),activation='relu',input_shape=(32,32,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(filters=16,kernel_size=(5,5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
```

```python
model.summary()
```

### Padding Argument takes 2 values:

    1. valid - Results in no padding
    2. Same - Results in padding with zeros and with strides = 1 ensures that the feature map is of the same size as that of the input

Padding is only w.r.t convolution layers. Default value for padding=valid & stride=1.

### Padding = 'valid'


```python
model=Sequential()
model.add(Conv2D(filters=6,kernel_size=(5,5),padding = 'valid', strides = 1, activation='relu',input_shape=(32,32,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(filters=16,kernel_size=(5,5),padding = 'valid', strides = 1,activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
```

```python
model.summary()
```

### Padding = 'same'


```python
model=Sequential()
model.add(Conv2D(filters=6,kernel_size=(5,5),padding = 'same', strides = 1, activation='relu',input_shape=(32,32,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(filters=16,kernel_size=(5,5),padding = 'same', strides = 1,activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
```

By controlling the strides parameter, you can adjust how much downsampling the MaxPooling2D layer performs, which in turn impacts the model's ability to generalize (potentially reducing overfitting) and the computational cost (fewer parameters and calculations in subsequent layers).






### Strides in Convolution and Max-pooling

Yes, the concept of strides in both Conv2D and MaxPooling2D layers is similar, but the operations being performed are different. In both cases, strides determine how the window (**be it a convolutional kernel or a pooling window**) moves across the input feature map. Specifically, the strides parameter specifies the step size to take along each dimension (height and width) when moving the window.



Summary:
In Conv2D, strides affects how the convolutional kernel moves when performing the convolution operation.
In MaxPooling2D, strides affects how the pooling window moves when performing the max-pooling operation.
Both serve to control the spatial dimensions of the output feature maps and, therefore, the number of operations that need to be computed, impacting both model performance and computational cost.

**Question**

Find out the value of p (padding) for:

    1. input of 32X32 with kernel of 5X5 resulting in an output of same size
    2. input of 16X16 with kernel of 5X5 resulting in an output of same size
    3. input of 16X16 with kernel of 3X3 resulting in an output of same size

**Solution1**

### $(n+2p)-k+1$

n = 32,
k = 5

32+2p-5+1 = 32

2p=4

p=2

**Solution2**

### $(n+2p)-k+1$

n = 16, k = 5

16+2p-5+1 = 16

2p=4

p=2

**Solution3**

### $(n+2p)-k+1$


n = 16, k = 3

16+2p-3+1 = 16

2p=2

p=1
