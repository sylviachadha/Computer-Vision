# Image Processing using Open CV

- Open CV is an open source computer vision library. It involves analysing images to produce useful information.
- **Open CV reads a given image in a BGR format by default**

## Aim

## Morphological Transformations

- Morphological Transformations based on the image shape.
- Generally performed on binary images.
- Needs 2 inputs, one is original image and 2nd is **structuring element or kernel** which decides the nature of operation.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
pwd
```

```python
img = cv2.imread('/Users/sylvia/Desktop/datasets/CV/opencv_inv.png')

plt.figure(figsize=(4,4))
plt.imshow(img);

img.shape
```

## <font color=blue> Erosion Transformation

### **Experiment with Structuring element / kernel** 

**Experiment 1, kernel1 = np.zeros**

```python
kernel1 = np.zeros((5,5), np.uint8)
kernel1
```

```python
erosion_img1 = cv2.erode(img, kernel1, iterations=1)
erosion_img1.shape
```

```python
plt.figure(figsize=(4,4))

plt.imshow(erosion_img1);
```

**Experiment 2, kernel2 = np.ones**

```python
kernel2 = np.ones((5,5), np.uint8)
erosion_img2 = cv2.erode(img, kernel2, iterations=1)
plt.figure(figsize=(4,4))
plt.imshow(erosion_img2);
```

The thickness of the size of the foreground object decreases when compared to the original image. This is called erosion.

**Experiment 3, kernel3 = random kernel**

Changing values or even size of kernel results same as Experiment 2??

```python
kernel3 = np.random.randint(0,255, size=(3,3))

kernel3
```

```python
erosion_img3 = cv2.erode(img, kernel3, iterations=1)
plt.figure(figsize=(4,4))
plt.imshow(erosion_img3);
```

```python
## Viewing all erosion plots together

plt.figure(figsize=(15,10))

plt.subplot(141)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(142)
plt.imshow(erosion_img1)
plt.title('Zeros kernel')

plt.subplot(143)
plt.imshow(erosion_img2)
plt.title('Ones Kernel')

plt.subplot(144)
plt.imshow(erosion_img3)
plt.title('Random Kernel');
```

## <font color=blue> Dilation Transformation

```python
dilation_img1 = cv2.dilate(img, kernel1, iterations=1)
dilation_img2 = cv2.dilate(img, kernel2, iterations=1)
dilation_img3 = cv2.dilate(img, kernel3, iterations=1)
```

```python
plt.figure(figsize=(15,10))

plt.subplot(141)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(142)
plt.imshow(dilation_img1)
plt.title('Zeros kernel')

plt.subplot(143)
plt.imshow(dilation_img2)
plt.title('Ones Kernel')

plt.subplot(144)
plt.imshow(dilation_img3)
plt.title('Random Kernel');
```
