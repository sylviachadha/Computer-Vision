# Viola-Jones algorithm 


- Haar cascade classifier, (a.k.a. Viola-Jones algorithm) is a very fast **face detection algorithm** that is used in most real time face recognition systems like digital cameras. Viola-Jones framework for object detection.
- Open CV contains many pre-trained classifiers for face, eyes, smile etc
- Viola-Jones first detects face on greyscale image and then finds location on colored image.

The primary aim of face detection algorithms is to determine whether there is any face in an image or not. Viola-Jones algorithm is an object-recognition framework that allows the detection of image features in real-time.

### Viola Jones Haar Cascades

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/Users/sylvia/Desktop/datasets/CV/om.jpg',1)
plt.imshow(img);    # By default BGR color code
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img);
```

```python
# Load Cascade classifier front face algorithm

face_cascade = cv2.CascadeClassifier('/Users/sylvia/Desktop/datasets/CV/haarcascade_frontalface_default.xml')
```

**.detectMultiScale**

```python
# Convert to greyscale and identify faces

gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray');
faces = face_cascade.detectMultiScale(gray_img,1.1,4)
faces
```

    **Docstring:**
    rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
    .   @brief Draws a simple, thick, or filled up-right rectangle.
    .   
    .   The function cv::rectangle draws a rectangle outline or a filled rectangle whose two opposite corners
    .   are pt1 and pt2.
    .   
    .   @param img Image.
    .   @param pt1 Vertex of the rectangle.
    .   @param pt2 Vertex of the rectangle opposite to pt1 .
    .   @param color Rectangle color or brightness (grayscale image).
    .   @param thickness Thickness of lines that make up the rectangle. Negative values, like #FILLED,
    .   mean that the function has to draw a filled rectangle.
    .   @param lineType Type of the line. See #LineTypes
    .   @param shift Number of fractional bits in the point coordinates.


```python
# Draw the faces

for x,y,w,h in faces:
    print(x)
    cv2.rectangle(rgb_img, (x,y), (x+w,y+h), (0,255,0), 3)
    
```
